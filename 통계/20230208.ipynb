```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


from matplotlib import font_manager, rc

font_path = './malgun.ttf'
font_name = font_manager.FontProperties (fname=font_path).get_name()
rc('font', family=font_name)

df1=pd.read_excel('시도_직업별_2008__이혼_20230227144800.xlsx')


df1=df1.fillna(method='pad')



df1.set_index(['시도별','직업별'],inplace=True)
s1= df1.columns.map(lambda x : x[:4])
s2 = df1.iloc[0].map(lambda x : x.split(' ')[0])
df1.columns=[s1,s2]

df1.columns.names = ['년도','구분']
df1 = df1.loc['전국']
df1=df1.astype(int)

list_a=[]
for i in range(0,len(df1.columns),1):
    if i % 2 == 0:
        df1.iloc[:,i] = df1.iloc[:,i] + df1.iloc[:,i+1]
        list_a.append(i+1)
df1.drop(df1.columns[[list_a]],axis=1,inplace=True)
df1.drop(df1.index[[9,10]],inplace=True)
df1.rename(columns={'남편':'이혼자 수'},inplace=True)
df1=df1.iloc[1:]
df1
```
